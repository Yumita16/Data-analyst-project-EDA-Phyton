Dari BOXPLOT dapat dilihat :
- Garis hijau mewakili median
- Kotak biru menunjukkan rentang IQR
- Garis horizontal menandakan bahwa data paling kecil (garis bawah) dan paling besar (garis atas)
- Tidak terdapat titik diluar garis : diatas angka 10 dan dibawah angka 1 menandakan bahwa tidak ada OUTLIER di kolom

```python
# deteksi outlier menggunakan IQR
df_ori3 = df_ori2.copy()

Q1=df_ori3['Compartments'].quantile(0.25)
Q3=df_ori3['Compartments'].quantile(0.75)
IQR=Q3-Q1

lower= Q1 - 1.5 * IQR
upper= Q3 + 1.5 * IQR

outlier_iqr = df_ori3[(df_ori3['Compartments'] < lower) | (df_ori3['Compartments'] > upper)]
print(outlier_iqr)
```

- Hasil dari running data didapatkan hasil bawah python memberitahu bahwa tabel tersebut kosong (tidak ada baris yang memenuhi kriteria sebagai pencilan)

### 4. Distribusi Data, mengecek normalisasi data

##### 1. Distribusi kolom numerik (Price)

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

df_ori4 = df_ori3.copy()

plt.figure(figsize=(5,3), facecolor='#f0f0f0') # Warna latar luar (abu muda)
ax = plt.axes()
ax.set_facecolor('#e6f2ff') # Warna latar dalam grafik (biru sangat muda)

# Brand
Price = np.array(df_ori4['Price'])

mu = np.mean(Price)
sigma = np.std(Price)

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = norm.pdf(x, mu, sigma)
plt.hist((df_ori4['Price']), bins=5, density=True, alpha=0.6, color="skyblue", edgecolor="black", align="left")
plt.plot(x, pdf)
plt.title("Histogram Data Price")
plt.xlabel("Value Price")
plt.ylabel("Frekuensi")
plt.show()
```

##### 1.1 Uji normalitas (uji Kolmogorov karena data >5000)

```python
df_ori4 = df_ori3.copy()

from scipy.stats import kstest, norm

data1 = df_ori4['Price']

stat, p = kstest(data1, 'norm', args = (np.mean(data1), np.std(data1)))
print("Statistik uji : ", stat)
print("p-value : ", p)

if p > 0.05:
    print("Data berdistribusi normal")
else :
    print("Data tidak berdistribusi normal") 
```

### 2. Menggunakan groupby() untuk melakukan agregasi (misal rata-rata penjualan per kategori).

##### 4. Analisis PMF

```python
# Menghitung PMF untuk kombinasi Material dan Size
# Tambahkan observed=False di dalam kurung groupby untuk versi baru kedepannya
pmf_data = df_stat.groupby(['Material', 'Size'], observed=False).size() / len(df_stat)
pmf_data = pmf_data.reset_index(name='Probability')

print("Tabel PMF Kombinasi Material & Size:")
display(pmf_data)
```

INSIGHT :
1. Hampir semua hasil data dari analisis PMF menunjukkan nilai probabilitynya 0.08 sekitar 8% yang menandakan jumlah datanya seimbang.
2. Setiap kombinasi antara material dan size memiliki peluang kemunculan yang hampir sama, tidak ada jenis tas yang mendominasi populasi data secara signifikan.
3. Hasil PMF menunjukkan keseimbangan kategori yang sangat baik dan akan menguntungkan untuk proses pemodelan berikutnya karena model tidak akan bias terhadap salah satu material atau ukuran tertentu.

##### 5. Uji statistik : kruskal-wallis test

```python
from scipy import stats

# Menyiapkan list untuk menampung grup data
groups = []

# Melakukan looping berdasarkan nilai unik di kolom 'Material'
for g in df_stat["Material"].unique():
    # Mengambil kolom 'Price' untuk setiap grup material dan menghapus NaN jika ada
    groups.append(df_stat[df_stat["Material"] == g]["Price"].dropna())

# Eksekusi Kruskal-Wallis Test
h_stat, p_val = stats.kruskal(*groups)

print("Kruskal-Wallis Test")
print("H-statistic:", h_stat)
print("p-value:", p_val)

# Batas signifikansi
alpha = 0.05

print("\nKesimpulan:")
if p_val < alpha:
    print("\nKesimpulan: Tolak H0 artinya terdapat perbedaan distribusi harga yang signifikan jika ditinjau dari jenis material dan ukuran tas.")
else:
    print("\nKesimpulan: Gagal Tolak H0 artinya tidak ada perbedaan distribusi harga yang signifikan jika ditinjau dari jenis material dan ukuran tas")
```

# 4. Teknik EDA Lanjutan

### 1. Teknik Groupby Aggregation

- Melakukan grouping berdasarkan dua kolom yaitu material dan size lalu menghitung nilai statistika harganya (mean, median, jumlah data).
- Hal ini dilakukan untuk melihat kombinasi tas yang paling mahal atau paling banyak di pasar

```python
# Grouping berdasarkan Material dan Size
df_group = df_final.groupby(['Material', 'Size'])['Price'].agg(['mean', 'median', 'count']).reset_index()

# Mengurutkan dari kombinasi yang rata-rata harganya paling mahal
df_group_sorted = df_group.sort_values(by='mean', ascending=False)
display(df_group_sorted)
```

INTERPRETASI :
- Setelah dianalisis secara interaktif atau dikombinasikan faktor jenis material dan ukuran tas terbukti tidak ada pengaruh terhadap harga jual tas sekolah di pasaran
- Kombinasi tas berbahan premium dan berukuran besar (seperti Leather-Large) akan bernilai lebih mahal tidak terbukti oleh data ini karena harga 81,2 USD setara dengan tas berbahan standar berukuran kecil (Polyester-small senilai 82.27 USD)

### 2. Teknik Cross Tabulation

- Melakukan teknik cross tabulation pada price
- Karena Price adalah data numerik kontinu, maka tidak bisa langsung memasukkannya ke crosstab biasa. Bisa menggunakan dengan menghitung rata-rata harga untuk setiap pertemuan antara Material (baris) dan Size (kolom) menggunakan argumen aggfunc.

```python
# Cross Tabulation untuk melihat rata-rata harga di setiap kombinasi
kombinasi_harga = pd.crosstab(
    index=df_final['Material'], 
    columns=df_final['Size'], 
    values=df_final['Price'], 
    aggfunc='mean' # menghitung rata-rata harga
).round(2)

display(kombinasi_harga)
```

```python
# visualisasi crosstab menggunakan heatmap untuk melihat hubungan
plt.figure(figsize=(6, 4))
sns.heatmap(kombinasi_harga, annot=True, fmt=".2f", cmap='YlGnBu', cbar=True)
plt.title('Matriks Rata-rata Harga: Material vs Size', fontsize=11, fontweight='bold', pad=10)
plt.show()
```

INTERPRETASI :
- Rata-rata harga tas di setiap titik pertemuan kategori bergerak sangat stabil dan rapat, yaitu hanya berkisar antara 80.21 USD hingga 82.98 USD. Selisih harga tertinggi dan terendah dari total 12 kombinasi unik ini tidak mencapai 3 USD.
- Hasil pengujian interaktif ini memperkuat temuan Advanced EDA pada groupby. Kombinasi karakteristik fisik (seperti tas Leather berukuran Large seharga 81.21 USD) tidak terbukti mendongkrak nilai jual produk menjadi lebih premium dibandingkan tas bermaterial standar dengan ukuran paling kecil (Polyester - Small seharga 82.27 USD)

### 3. Teknik Filtering

##### a. Melakukan filter data berdasarakn material dan size untuk melihat apakah termasuk kelompok harga mahal

- Melakukan filtering untuk memisahkan tas sekolah yang berada di kelompok harga mahal, harga murah.
- Setelah difilter lalu dicek kelompok tas mahal yang didominasi oleh material dan ukuran 

```python
# Menentukan batasan harga mahal (misalnya tas di atas 110 USD atau menggunakan percentile 85)
batas_mahal = df_final['Price'].quantile(0.85)

# Pertama memfilter data tas yang harganya mahal
df_mahal = df_final[df_final['Price'] >= batas_mahal]

# Melakukan cek distribusi material dan ukuran di kelompok tas mahal tersebut
print(f"--- Distribusi Material pada Tas Mahal (>= {batas_mahal:.2f} USD) ---")
print(df_mahal['Material'].value_counts(normalize=True) * 100)

print(f"\n--- Distribusi Ukuran pada Tas Mahal (>= {batas_mahal:.2f} USD) ---")
print(df_mahal['Size'].value_counts(normalize=True) * 100)
```
